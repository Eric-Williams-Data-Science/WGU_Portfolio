```python
#!/usr/bin/env python
# coding: utf-8
import pandas as pd
df = pd.read_excel(r"C:\Users\18014\Desktop\D598 Data Set.xlsx")

#Instruction: Identify duplicate rows
duplicate_rows = df[df.duplicated()]

duplicate_rows
```

```python
#There appear to be no diplicate rows!

#Instruction: Group by business state, calculate mean, median, min, and max of each numerical column
#Use aggregate to run several descriptive statistics, store as new dataframe df2

df2 = df.groupby("Business State").agg(
    Total_Long_term_Debt_mean=("Total Long-term Debt", "mean"),
    Total_Long_term_Debt_median=("Total Long-term Debt", "median"),
    Total_Long_term_Debt_min=("Total Long-term Debt", "min"),
    Total_Long_term_Debt_max=("Total Long-term Debt", "max"),
    Total_Equity_mean=("Total Equity", "mean"),
    Total_Equity_median=("Total Equity", "median"),
    Total_Equity_min=("Total Equity", "min"),
    Total_Equity_max=("Total Equity", "max"),
    Debt_to_Equity_mean=("Debt to Equity", "mean"),
    Debt_to_Equity_median=("Debt to Equity", "median"),
    Debt_to_Equity_min=("Debt to Equity", "min"),
    Debt_to_Equity_max=("Debt to Equity", "max"),
    Total_Liabilities_mean=("Total Liabilities", "mean"),
    Total_Liabilities_median=("Total Liabilities", "median"),
    Total_Liabilities_min=("Total Liabilities", "min"),
    Total_Liabilities_max=("Total Liabilities", "max"),
    Total_Revenue_mean=("Total Revenue", "mean"),
    Total_Revenue_median=("Total Revenue", "median"),
    Total_Revenue_min=("Total Revenue", "min"),
    Total_Revenue_max=("Total Revenue", "max"),
    Profit_Margin_mean=("Profit Margin", "mean"),
    Profit_Margin_median=("Profit Margin", "median"),
    Profit_Margin_min=("Profit Margin", "min"),
    Profit_Margin_max=("Profit Margin", "max")
)

#Test Data Frame 2 to inspect it
df2
```

```python
#Instruction: Filter original df to identify all businesses with debt-to-equity ratios that are negative.
negative_debt_to_equity_ratio = df[df["Debt to Equity"] <0]

#Check negative_debt_to_equity_ratio to test it
negative_debt_to_equity_ratio
```

```python
#Instruction: Create a new data frame that provides the debt-to-income ratio for every business in the data set
#Note: Debt-to-income ratio is defined as long-term debt divided by revenue.

df["Debt to income ratio"] = df["Total Long-term Debt"] / df["Total Revenue"]

#Instruction: Concatenate the debt-to-income ratio data frame you created with the original data frame
df_final = pd.concat([df, df["Debt to income ratio"]], axis=1)


#Check to see if debt to income ratio appears properly in df_final
df_final
```
